# Analysis using Gemini 1.5 Pro

This notebook summarises the results of a quick analysis with Gemini 1.5 Pro, done via [Google AI Studio](https://aistudio.google.com/) on 2024-08-28. The core of the analysis was done via the studio interface for quicker iteration and to reduce costs, while taking advantage of Gemini 1.5 Pro's large context window.

## Preprocessing

The following steps were run on Ubuntu 22.04:

1. Download all the available PDFs to `raw_docs/` (up to submission 195 at the time of the analysis)
2. In `raw_docs/`, convert all the PDFs to text files:

   ```bash
   $ for file in *.pdf; do pdftotext "$file" "${file%.pdf}.txt"; done
   ```

3. Concatenate the text files to a single file, delimiting each file's contents with START/END headings:

    ```bash
    $ for file in Sub*.txt; do
      echo "START $file"
      echo "---"
      cat "$file"
      echo "---"
      echo "END $file"
    done > concatenated_output.txt
    ```

4. Convert `concatenated_output.txt` to PDF for upload to Gemini: Print to file from the text editor without page editing.

This results in a single file with about 490k Gemini tokens (well within the 1M context window).

````python
def parse_classification_bullets(model_output: str, name: str) -> pd.Series:
    sub_nums = []
    sub_classes = []
    for row in model_output.split("\n"):
        if not row.startswith("- Sub"):
            if row.strip():
                print(f"[Skipped] {row}")
            continue
        sub_num, sub_cls = row.strip().replace("- Sub", "").split(": ")
        sub_nums.append(int(sub_num))
        sub_classes.append(True if sub_cls == 't' else False if sub_cls == 'f' else None)
    result = pd.Series(sub_classes, sub_nums, name=name)
    print(f"[Summary for '{name}'?] True: {result.sum()}. False: {(result == False).sum()}. Unknown: {result.isna().sum()}")
    return result

jews = parse_classification_bullets(FULL_CHAT_HISTORY[-5]["parts"][0], name="jewish")
jews
````
